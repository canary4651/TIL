# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11) #0-5까지 11개의 데이터 줄세우기
y = x**2
```

```python
plt.plot(x,y)
plt.title("y=x^2")
plt.xlabel("ind var")
plt.ylabel("dep var")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("y=x^2")
plt.subplot(1,2,2)
plt.plot(y,x)
plt.title("y=x^(o.5)")
```

```python
plt.subplot(2,1,1) #행과 열, 그리고 index (첫번째) 
plt.plot(x,y)
plt.subplot(2,1,2) #행과 열, 그리고 index (두번째) 
```

# 객체지향 프로그래밍 -object oriented method

```python
# 새로운 figure를 만든다 (캔버스)
# 그 figure에다가 새로운 axes를 한다 (그림을 그린다)
# 위에서는 그냥 plot에서 바로 그림을 그려서 보여줬는 데 
# object oriented method는 figure를 우리가 만들고, 언제든지 나중에 불러올 수 있음 (남아있음 재사용하는 오브제를 만든거임)
```

```python
fig = plt.figure() #캔버스 먼저 그리기 
axes = fig.add_axes([0,0,1,1]) #axex는 위치를 설정해줘야함 왼쪽,오른쪽,가로,세로 
axes # 이것만 찍어보면 <matplotlib.axes._axes.Axes at 0x1117a7610> 고유번호를 확인할 수 있음 
axes.plot(x,y)
axes.set_title("y=x^2") 
axes.set_xlabel("ind var")
axes.set_ylabel("dep var")
```

```python
# 위치 예시 (fig.add 안의 []설명 )
fig=plt.figure()
axes1=fig.add_axes([0.5,0,1,1]) 
axes2=fig.add_axes([0,0.5,1,1])
```

```python
#작은 plot을 큰 plot 안에 넣기 
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1]) 
axes2=fig.add_axes([0.1,0.5,0.5,0.4])
#안에 있는 게 exes2
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title("y=x^(0.5)")
axes1.set_title("y=x^2")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
```

# 객체지향 방법에서 subplot 쓰기 
# 특이한 점 : 생성했을 때 결과물이 튜플형태(,)로 나누어져 있음 

```python
# 객체지향 방법에서 subplot 쓰기 
# 특이한 점 : 생성했을 때 결과물이 튜플형태(,)로 나누어져 있음 
fig, axes = plt.subplots(1,2) # 행,열 #튜플 형태가 나오기 때문에 각 객체의 이름을 지정할 수 있음 
# object 자체는 fig(canvas)
# axes는 안의 내용 진짜 axes(그림그리기)
```

```python
axes[0].plot(x,y)
axes[1].plot(y,x) #축에다가 명령을 실행 
fig
```

# 범례지정 (legend)

```python
# 범례지정 (legend)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="y=x^2")
ax.plot(x,x**3,label="y=x^3") #같은 그래프에 2개 그리기 
# 이때 각각 그래프가 뭘 의미하는 지 넣어주는 게 legend이다(3번째에 넣음) - 근데 불러내지 않는 이상 호출은 안됨 
ax.legend() #요렇게 해줘야 불러져짐 - 위치도 조정가능 
# ax.legend(loc = 0 ) # 위치 조정가능 (0,1,2,3,4 하나씩 넣어서 바꿀 수 있음) 근데 안하면 자신이 알아서 최적의 위치를 찾아서 넣어줌 
# 좌표로 갈 수도 있음 ax.legend(loc = [0.1, 0.2] ) 이렇게..! 밖으로 낼 수도 있음 
```

```python
import pandas as pd
import seaborn as sns
# matplotlib 도 불러오고 인라인도 해주기 - 다른 파일 만들 때
```

```python
sns.set()
```

```python
tips =sns.load_dataset("tips") #씨본 자체의 데이터셋
```

```python
tips.head()
```

```python
sns.relplot(x ='total_bill', y='tip', data =tips , col='time',
            hue='smoker', style = 'smoker', size='size') # x, y 값 넣기  
# col = 시간대별로 보여줘 hue=흡연자를 색으로 구분해줘
# style: 안하면 그냥 다 동그라미 마크인데 이거하면 흡연자를 다른 모양으로 바꿔줌 size : 크기 별로 보여줌 
```

```python
# relation plot을 line plot으로 보기 
time = np.arange(500)
value = np.random.randn(500).cumsum() #누적합
```

```python
df=pd.DataFrame(dict(time=time, value = value))
sns.relplot(x='time',y='value', kind='line' ,data=df) 
# 시계열의 형태는 라인으로 보는 게 더 보기쉬움 (kind=lind) - kind defalt 값은 산점도임
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', kind ='line' , data=fmri)
# 위에서 그냥 점 데이터였는 데 어떻게 라인이 나오냐? -> 임의로 중간정도 값으로 추정되는(estimate) 부분을 만들어서 라인으로 연결해줌 
# -> 회귀 함수를 그렸다고 생각하면 됨 (평균값과 신뢰구간(진한 면적)
```

```python
sns.relplot(x='timepoint', y='signal', kind ='line', hue ='event', style='event', ci= None, data=fmri)
# estimator=None (추정된 값들을 빼고 보여줌) #ci= None (confidenct 신뢰구간) : 위에서 shadding된 부분이 없어짐
```
