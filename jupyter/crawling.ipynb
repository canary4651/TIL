```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('/Users/Dahye/Desktop/notebooks/chromedriver')
#     3   -    
driver.implicitly_wait(3)
# url 
driver.get('https://datalab.naver.com/keyword/realtimeList.naver?where=main')
#   BeautifulSoup  
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

#   requests BeautifulSoup    
rank_list = soup.select(".rank_list.v2")[0].select('.list_area')
rank_result_list = []
for rank in rank_list:
    # print(rank.text.split())
    # split()       .
    #  , ['1', '879']    ,   .
    rank_result = rank.text.split()
    
    #        Dictionary  .
    rank_result_list += [{"rank" :rank_result[0], "value": rank_result[1]}]
print(rank_result_list)
print("-"*100)
print('   {} "{}".'.format(rank_result_list[0]["rank"], rank_result_list[0]["value"]))
```
