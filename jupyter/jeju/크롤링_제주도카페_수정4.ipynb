```python
cafes = []
menus = []
comments = []
for id in ids:
    res = requests.get(f'https://place.map.kakao.com/main/v/{id}')
    data = res.json()
    try:
        cafes.append({
            'cafe_id': id,
            'cafe': data['basicInfo']['placenamefull'],
            'address': data['basicInfo']['address']['region'].get('fullname',''),
            'addrbunho' : data['basicInfo']['address']['addrbunho'],
            'image_url': data['basicInfo'].get('mainphotourl', ''),
        })
    except:
        cafes.append({
            'cafe_id': id,
            'cafe': data['basicInfo']['placenamefull'],
            'address': data['basicInfo']['address']['region'].get('fullname',''),
            'addrbunho' : '',
            'image_url': data['basicInfo'].get('mainphotourl', ''),
        })
        
    # menu
    if 'menuInfo' in data:
        for menu in data['menuInfo']['menuList']:
            menus.append({
                'cafe_id': id,
                'cafe': data['basicInfo']['placenamefull'],
                'name': menu['menu'],
                'price': menu.get('price', ''),
            })
    # comment
    for page in range(1, ceil(data['comment']['allComntcnt'] / 5) + 1):
        res = requests.get(f'https://place.map.kakao.com/commentlist/v/{id}/{page}')
        data = res.json()
        for comment in data['comment']['list']:
            comments.append({
                'cafe_id': id,
                'cafe': data['comment']['placenamefull'],
                'user_id': comment.get('commentid',''),
                'message': comment.get('contents', ''),
                'point': comment.get('point', ''),
                'likeCnt': comment.get('likeCnt', ''),
                'date': comment.get('date', '')
                })

            
pd.DataFrame(cafes)
```
