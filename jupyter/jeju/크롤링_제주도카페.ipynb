```python
ids = [
    '26094495',
    '819891458',
    '1852395758',
]
```

```python
from math import ceil

import requests
from bs4 import BeautifulSoup

cafes = []
menus = []
comments = []

for id in ids:
    res = requests.get(f'https://place.map.kakao.com/main/v/{id}')
    data = res.json()
    cafes.append({
        'id': id,
        'name': data['basicInfo']['placenamefull'],
        'image_url': data['basicInfo']['mainphotourl'],
    })

    # menu
    if 'menuInfo' in data:
        for menu in data['menuInfo']['menuList']:
            menus.append({
                'cafe_id': id,
                'cafe': data['basicInfo']['placenamefull'],
                'name': menu['menu'],
                'price': menu['price'],
            })

    # comment
    for page in range(1, ceil(data['comment']['allComntcnt'] / 5) + 1):
        res = requests.get(f'https://place.map.kakao.com/commentlist/v/{id}/{page}')
        data = res.json()
        for comment in data['comment']['list']:
            comments.append({
                'cafe_id': id,
                'cafe': data['comment']['placenamefull'],
                'message': comment.get('contents', ''),
                'point': comment.get('point', ''),
            })

pd.DataFrame(cafes)
```
