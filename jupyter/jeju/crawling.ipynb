```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('/Users/Dahye/Desktop/notebooks/chromedriver')
# 페이지가 모두 로드될 때까지 3초 기다려주는 코드 - 자바스크립트를 제대로 읽을때까지 가다림
driver.implicitly_wait(3)
# url에 접근
driver.get('https://datalab.naver.com/keyword/realtimeList.naver?where=main')
# 소스를 읽어 BeautifulSoup를 이용해 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 그다음은 이전에 requests와 BeautifulSoup를 이용했던 코드와 동일하게 처리
rank_list = soup.select(".rank_list.v2")[0].select('.list_area')
rank_result_list = []
for rank in rank_list:
    # print(rank.text.split())
    # split()을 이용해 필요한 값들을 나누어 리스트 형태로 만든다.
    # 미리 확인해본대로, ['1', '로또879당첨번호']와 같이 리스트의 첫번째에는 랭킹, 두번째에는 검색어가 나온다.
    rank_result = rank.text.split()
    
    # 이것을 우리가 이용하기 쉬운 형태로 만들어주기 위해 Dictionary 형태로 구성한다.
    rank_result_list += [{"rank" :rank_result[0], "value": rank_result[1]}]
print(rank_result_list)
print("-"*100)
print('실시간 급상승 검색어 {}위는 "{}"입니다.'.format(rank_result_list[0]["rank"], rank_result_list[0]["value"]))
```
