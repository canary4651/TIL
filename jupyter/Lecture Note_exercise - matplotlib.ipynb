# matplotlib - Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 100)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])

ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20, 22)
ax2.set_ylim(350, 550)
```
