```python
a1 = [ '지번주소', '좌표정보', '좌표정보.1','콘텐츠 명' ]
aa = a[a1]
```

```python
# 지도 준비

from folium import Map
from folium.plugins import HeatMap
map = Map(location=[36.5, 127.5], zoom_start=7)



coordinates = aa[['좌표정보.1', '좌표정보']] #일자리카페는 위도 경도 칼럼이 바껴져 있음! 

coordinates
```

```python
# 지도 준비

map = Map(location=[36.5, 127.5], zoom_start=7)

# 지도에 위치 표시

for lat, lng in coordinates.values:
    marker = folium.Circle(
        location=[lat, lng],
        radius=600,
        weight=1,
        fill_color='#F00', #RGB
        fill_opacity=0.3, #투명도 
    )
    marker.add_to(map)

map
map.save(outfile='노인일자리카페.html')
```

```python
map = Map(location=[36.5, 127.5], zoom_start=7)

# 히트맵 표시

HeatMap(data=coordinates, radius=20).add_to(map)

map
```
