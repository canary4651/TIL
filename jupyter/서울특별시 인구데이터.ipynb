```python
import pandas as pd
```

```python

population = pd.read_csv('TB_T_RSPOP_ADMI.txt', sep='|')
population
```

```python
population.info()
```

```python
population.nunique()
```

```python
population.columns = ['기준년월', '행정명코드', '행정동명', '성별코드','연령대코드','거주인구수','시군구명']
population
```

```python
sp = population[['시군구명','행정동명','거주인구수']] 
sp   # simple population
```

```python
sp['sigungu'] = sp.iloc[:, 0].str[1:-1]
```

```python
sp['dong'] = sp.iloc[:, 1].str[1:-1]
```

```python
sp['nop'] = sp.iloc[:, 2].str[1:-1]  # nop: number of people
```

```python
sp = sp.drop(columns=['시군구명', '행정동명', '거주인구수'])
```

```python
sp.to_csv('빅캠 인구수.csv')
```
