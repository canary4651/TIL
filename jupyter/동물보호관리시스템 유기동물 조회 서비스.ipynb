# ElementTree로 XML 데이터 파싱
- https://docs.python.org/3/library/xml.etree.elementtree.html

````python
import xml.etree.ElementTree as ET
# XML 데이터 파싱 parsing 데이터를 조립해 원하는 데이터를 빼내는 프로그램을 하는것
root = ET.fromstring(response.text) #데이터(위에서 찾은)를 tree화 시킴! (우리가 활용하기 좋은 형태로 바꿈)
# item 목록에 있는 kindCd와 specialMark 항목 출력

for item in root.iter('item'): #iter를 써야 'item'이 나올 때까지 밑으로 내려가서 찾음 () 안의 내용을 찾음
    kind = item.find('kindCd').text #cd = code 약자 #<> 얻을 때는 tag, 그옆의 내용을 얻을 때는 text 
    description = item.find('specialMark').text #내가 원하는 정보 넣으면 됨
    print(f'{kind} | {description}') # | 넣어서 보여주라는 말 #f-string <- fommating의 방법 
````

# 컬럼 목록 얻기
ElementTree는 Iterator를 사용합니다.

next 함수를 사용하면 for를 쓰지 않고도 직접 값을 얻을 수 있습니다.

Iterator에 대한 파이썬 위키의 설명: https://wiki.python.org/moin/Iterator

````python
# 첫번째 item을 얻습니다
item = next(root.iter('item')) 
#next쓰면 for문 안쓰고도 for문을 돌린 것처럼 할 수 있음
#iterator : 최초의 것을 얻은 후 그 다음, 그 다음, 그 다음 (for문이 비슷한 방식으로 작동함)
# next는 iterator 에서 쓸 수 있음 
# '예시' a = enumerate[10, 20, 30]
# next(a) 해주면 (0, 10) 한번더 next(a) 해주면 (1,20) 한번더 (2,30) 한번 더 (3, error)- 그 다음은 없으니까 
item
````

````python
#item 바로 아래에 뭐가 있을까? 
list(item)
````

````python
# 컬럼 목록 준비 
columns = []
#Accumlation
for child in list(item): #tree에서 위를 부모 parent라고 부르고 밑을 자녀 child 라고 부른다
    columns.append(child.tag) #컬럼 이름만 뽑고 싶을 때 
#컬럼 목록 확인 
print(columns)
````

# XML 데이터를 Pandas로 변환
https://pandas.pydata.org/

````python
# pandas 사용
import pandas as pd
pd.__version__
````

````python
# XML 데이터를 dictionary의 list로 만들기
rows = [] # 빈 리스트 준비 
for item in root.iter('item'): #for문 
    row = {} #빈 딕셔너리 준비 
    for child in list(item):
        row[child.tag] = child.text #child안에 있는 값 # [키] = value 
    rows.append(row)

print('갯수:', len(rows))
print()
print(rows[0]) #0번째만 보기 
````

````python
# DataFrame으로 만들기

df = pd.DataFrame(rows, columns=columns)

df
````

````python
data = df['age'].value_counts()

pd.DataFrame(data)
````
