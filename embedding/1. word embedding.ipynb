### Skip-Gram 학습데이터 만들기

Skip-Gram은 `타겟 단어(target word)`로 `주변 단어(context word)`를 맞추는 과정에서 학습하는 모델입니다. `(타겟 단어, 주변 단어)` 쌍을 만들어 봅시다.

```python
def generate_skipgram_training_data():
    training_data = []
    for sentence in corpus:
        sent_len = len(sentence)
        for i, word in enumerate(sentence):
            w_target = sentence[i]
            for j in range(i - settings['window_size'], i + settings['window_size'] + 1):
                if j != i and j <= sent_len - 1 and j >= 0:
                    w_context = sentence[j]
                    training_data.append([w_target, w_context])
    return training_data
```

```python
train_data = generate_skipgram_training_data()
```

## Forward Computation

Skip-Gram은 타깃 단어($t$)로 문맥 단어($c$)를 맞추는 과정에서 학습됩니다.

𝐔 는 타겟 단어, 𝐕는 주변 단어와 관련한 행렬입니다.

$\mathbf{u_{t}} = \text{onehot}_{t} \cdot \mathbf{U} \\
\text{logit} = \mathbf{u_{t}} \cdot \mathbf{V} \\
P(c|t) = \text{softmax(logit)}$
- 원 핫 벡터를 구하면 행렬의 타겟 워드만 나오는 참조 현상이 나옴 ( 사진 참고 )

```python
def forward_computation(target_word_onehot_vectors, parameters):
    # target_word_onehot_vectors shape : [batch_size, vocab_size]
    # U shape : [vocab_size, settings["dimension"]]
    # target_word_vectors shape : [batch_size, settings["dimension"]]
    target_word_vectors = np.dot(target_word_onehot_vectors, parameters["U"])
    # logits shape : [batch_size, vocab_size]
    logits = np.dot(target_word_vectors, parameters["V"])
    # 0.0001 : underflow 방지
    softmax_probs = softmax(logits) #forword computation 
    #softmax_probs = np.divide(np.exp(logits), np.sum(np.exp(logits), axis=0, keepdims=True) + 0.001)
    if len(target_word_vectors.shape) == 1:
        target_word_vectors = np.expand_dims(target_word_vectors, axis=0)
    # 나중에 backprop을 하기 위한 데이터 캐시
    caches = {"target_word_vectors": target_word_vectors}
    return softmax_probs, caches
```

## softmax 함수 정의

$\exp$는 1보다 크기 때문에 지수승을 거듭하면 지나치게 큰 수가 나옴

이에 입력 벡터 $x$의 최대값을 $x$에서 빼줌

이렇게 하더라도 $x$ 벡터 요소들의 대소관계는 바뀌지 않음

```python
np.exp(2000)
```

```python
def softmax(x, axis=1):
    x = x - x.max(axis=axis, keepdims=True) # x에서 최대값 x를 빰 (대소 관계가 바뀌지 않음!!!)
    y = np.exp(x)
    return y / y.sum(axis=axis, keepdims=True)
```

## Loss Computation

```python
def cross_entropy(target, softmax_prob): #target = 워드랑 헷갈라지 말지. label(정답), 예측한 값 -> 이 둘의 차이가 cross entropy
    """
    targe: shape (batch_size, vocab_size)
    softmax_prob: output out of softmax. shape: (batch_size, vocab_size)
    """
    batch_size = softmax_prob.shape[0]
    cost = -(1 / batch_size) * np.sum(np.sum(target * np.log(softmax_prob + 0.001), axis=0, keepdims=True), axis=1)
    return cost
```

## Back Propagation

https://ratsgo.github.io/deep%20learning/2017/10/02/softmax/

```python
def softmax_backward(target, softmax_prob):
    """
    target: labels onehot vector of training data. shape: (batch_size, vocab_size)
    softmax_prob: output out of softmax. shape: (batch_size, vocab_size)
    """
    return softmax_prob - target
# 최초의 그레이던트를 구하는 방법 : softmax_prob에서 정답(target)을 빼누다 
```

```python
def dense_backward(dL_dlogits, caches):
    V = parameters['V']
    batch_size = dL_dlogits.shape[0]
    target_word_vectors = caches['target_word_vectors']
    dL_dtarget = (1 / batch_size) * np.dot(dL_dlogits, V.T)
    dL_dV = np.dot(dL_dlogits.T, target_word_vectors)
    return dL_dtarget, dL_dV
# 2,3번 구하는 거. ut와 v의 상대 로컬 곱하는 거 
```

## parameter update

$\mathbf{u}_{t}$는 $\mathbf{U}$ 행렬에서 타겟 단어(t)에 해당하는 행 벡터를 참조(look)한 것

따라서 $U$ 행렬의 t에 해당하는 행벡터만 업데이트함
- v는 한번 업뎃할 때마다 행렬 전체가 업뎃 됨
- ut는 그 타겟만 업뎃됨 

```python
def update_parameters(parameters, dL_dtarget, dL_dV,
                      learning_rate, target_words_idexes):
    parameters['V'] -= learning_rate * dL_dV.T #v는 한번 업뎃할 때마다 행렬 전체 업뎃(덮어씀)
    if type(target_words_idexes) == list: # 행벡터 하나씩 업뎃하는 거라서 for문을 돌림 
        for target_word_idx, target_word_update in zip(target_words_idexes, dL_dtarget):
            parameters['U'][[target_word_idx], :] -= learning_rate * target_word_update
    else:
        parameters['U'][[target_words_idexes], :] -= learning_rate * dL_dtarget
```

```python
parameters = {} #로스가 100 에폭(epochs)마다 한번씩 돌아가게 바꿈 # 정해진 에폭스가 되었을 때나 목표값이 나올 때 중단시킴.
parameters["U"] = np.random.randn(vocab_size, settings["dimension"]) * 0.01 #행렬 만들기
parameters["V"] = np.random.randn(settings["dimension"], vocab_size) * 0.01
target_word_onehot_vectors = [word2onehot(target_word) for target_word, context_word in train_data] #원핫 만들기
context_word_onehot_vectors = [word2onehot(context_word) for target_word, context_word in train_data]
target_words_idexes = [word_to_index[target_word] for target_word, context_word in train_data] #인덱스 주기 
for i in range(settings["epochs"]): #순방향 계산
    softmax_prob, caches = forward_computation(target_word_onehot_vectors, parameters) #로스 계산
    loss = cross_entropy(context_word_onehot_vectors, softmax_prob) #로스 최소화
    dL_dlogits = softmax_backward(context_word_onehot_vectors, softmax_prob) #로스를 백프롭으로 계산함
    dL_dtarget, dL_dV = dense_backward(dL_dlogits, caches)
    update_parameters(parameters, dL_dtarget, dL_dV, settings['learning_rate'], target_words_idexes) #업데이트 시킴
    if i % 100 == 0: 
        print(loss)
```

```python
# 코사인 시밀러리티 구하기 (공책 설명 - 벡터 설명 보기) => 유닛 벡터 내적함을 보여주기 위해 
def most_similar(query_word, topn=3, embeddings=parameters["U"]):
    # embeddings shape : [vocab_size, dimension]
    embedding_norms = np.linalg.norm(embeddings, axis=1)
    query_vector = embeddings[word_to_index[query_word]] / np.linalg.norm(embeddings[word_to_index[query_word]])
    cosine_simiarities = []
    for word_idx, (word_vector, norm) in enumerate(zip(embeddings, np.linalg.norm(embeddings, axis=1))): #norm 은 벡터의 기본 단위
        cosine_similarity = np.dot(query_vector, word_vector / norm) # 코사인 시밀리리티를 높은 순서대로 소팅해서 보여주는 함수 식
        cosine_simiarities.append([index_to_word[word_idx], cosine_similarity])
    return sorted(cosine_simiarities, key=lambda x: x[1], reverse=True)[1:topn+1]
```
